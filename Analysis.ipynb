```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('NBA_MVP_votegetters_data.csv')
mvps = data[data['MVP']==1]
rest = data[data['MVP']==-1]
```

```python
print(len(rest.columns))
rest.columns
```

```python
fig = plt.figure(figsize = (15,40))

for i in range(1, 46):
    plt.subplot(9,5,i)
    category = data.columns[i+7]
    plt.title(category)
    plt.bar(['MVPs', 'Rest'], [mvps[category].mean(), rest[category].mean()], width=0.25, align = 'center')
    plt.xlim(-1,2)
    
fig.savefig('mvps_vs_rest.png')
```

```python
pg = data[data['Pos']=='PG']
sg = data[data['Pos']=='SG']
sf = data[data['Pos']=='SF']
pf = data[data['Pos']=='PF']
c = data[data['Pos']=='C']

mvps_pg = mvps[mvps['Pos']=='PG']
mvps_sg = mvps[mvps['Pos']=='SG']
mvps_sf = mvps[mvps['Pos']=='SF']
mvps_pf = mvps[mvps['Pos']=='PF']
mvps_c = mvps[mvps['Pos']=='C']

rest_pg = rest[rest['Pos']=='PG']
rest_sg = rest[rest['Pos']=='SG']
rest_sf = rest[rest['Pos']=='SF']
rest_pf = rest[rest['Pos']=='PF']
rest_c = rest[rest['Pos']=='C']

f1 = plt.figure(figsize=(12,5))
plt.title('Positions of Votegetters')
plt.bar(['PG', 'SG', 'SF', 'PF', 'C'], [len(pg), len(sg), len(sf), len(pf), len(c)], width=0.3)
f1.savefig('positions_of_votegetters.png')
```

```python
f2 = plt.figure(figsize=(12,5))
plt.title('Positions of MVPs')
plt.bar(['PG', 'SG', 'SF', 'PF', 'C'], [len(mvps_pg), len(mvps_sg), len(mvps_sf), len(mvps_pf), len(mvps_c)], width=0.3)
f2.savefig('positions_of_mvps.png')
```

```python
import numpy as np

pg_count = []
sg_count = []
sf_count = []
pf_count = []
c_count = []

for year in range(1980, 1990):
    yeardata = data[data['Year']==year]
    pg_count.append(len(yeardata[yeardata['Pos']=='PG']))
    sg_count.append(len(yeardata[yeardata['Pos']=='SG']))
    sf_count.append(len(yeardata[yeardata['Pos']=='SF']))
    pf_count.append(len(yeardata[yeardata['Pos']=='PF']))
    c_count.append(len(yeardata[yeardata['Pos']=='C']))

labels = range(1980, 1990)
locations = np.arange(len(labels))
width = 0.15

fig1, ax = plt.subplots(figsize=(20,10))
rects1 = ax.bar(locations - width*2, pg_count, width, label='PG', color='y')
rects2 = ax.bar(locations - width, sg_count, width, label='SG', color='g')
rects3 = ax.bar(locations, sf_count, width, label='SF', color='r')
rects4 = ax.bar(locations + width, pf_count, width, label='PF', color='m')
rects5 = ax.bar(locations + width*2, c_count, width, label='C')

ax.set_ylabel('Counts')
ax.set_title('Counts by Year and Position: 1980-1989')
ax.set_xticks(locations, labels)
ax.legend()

fig1.savefig('1980-1989.png')
```
